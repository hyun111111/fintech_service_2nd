# 전체 페이지 수집하기 

```python
page = 1
final_result = []
while page < 9 :
    url = "https://www.yes24.com/product/category/bestseller"
    payload = dict(CategoryNumber="001", pageNumber=page, pageSize=120)
    r = requests.get(url,params=payload)
    print(r.url, r.status_code)
    soup = bs(r.content,"lxml")
    time.sleep(5)
    
    result = {}
    book_list = soup.select("ul#yesBestList > li")
    for idx, book in enumerate(book_list):
        print(f"{idx}/{len(book_list)} 추출중", end="\r")
        book_title = book.select_one("a.gd_name").text
        book_au = book.select_one("span.authPub.info_auth a").text if book.select_one("span.authPub.info_auth a") != None else " "
        book_p = book.select_one("span.authPub.info_pub a").text if book.select_one("span.authPub.info_pub a") != None else " "
        book_price = book.select_one(".info_row.info_price em.yes_b").text.replace(",","")
        book_rating = book.select_one(".rating_grade em.yes_b").text if book.select_one(".rating_grade em.yes_b") != None else 0
        book_em = book.select_one(".info_row.info_rating em.txC_blue").text if book.select_one(".info_row.info_rating em.txC_blue") != None else 0
        link = "http://www.yes24.com"+ book.select_one("a.gd_name")['href']
        
        key = ['book_title','book_au','book_p','book_price', 'book_rating','book_em']
        value = [book_title, book_au, book_p, book_price, book_rating , book_em]
        for key, value in zip(key, value):
            result.setdefault(key, []).append(value)
    final_result.append(pd.DataFrame(result))
    page += 1

final_result = pd.concat(final_result)
final_result = final_result.reset_index(drop=True)
final_result
```
