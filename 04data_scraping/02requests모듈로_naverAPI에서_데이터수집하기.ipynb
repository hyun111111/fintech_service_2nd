```python
def json_unpack(data):
    result = {}
    for page in data['items'] :
        for key, value in page.items():
            result.setdefault(key,[]).append(value)
    import pandas as pd
    result2 = pd.DataFrame(result)
    return result2
```

# 네이버 API로 검색한 결과 최대한 가져오기
* 1회 네이버 서버에 요청
* 응답에서 total을 찾아서 display로 나누어 총 반복할 페이지 산출
* 반복문 작성 및 종료조건 설정
* 1페이지 결과를 데이터 프레임으로 만들고 리스트에 append
* 반복이 끝난 후 리스트에 append된 데이터프레임을 1개로 합침

```python
search = input('책을 검색해주세요')
page = 0
data_set = []
while True :
    url = "https://openapi.naver.com/v1/search/book.json"
    payload = {'query': search ,'display': 100, 'start': (page*100)+1, 'sort': "sim" }
    headers = {"X-Naver-Client-Id" : "ky2xDsaZyifHgJ6pdBwU", "X-Naver-Client-Secret": "3nNHDxyhXR"}
    r = requests.get(url, params=payload, headers=headers)
    data = r.json()
    print(r.url)
    print(r.status_code)
    print("전체 검색 테이터수:", data['total'], "검색 페이지수:", f"{page}/{int(data['total']//100)}")
    if page == 10:
        break
    elif page < int(data['total']//100) :
        resurt_unpack = json_unpack(data)
        data_set.append(resurt_unpack)
        print(data_set[page])
        page += 1
    elif page == 10:
        page += 0.99
```

```python
final_result = pd.concat(data_set)
final_result = final_result.reset_index(drop=True)
final_result
```

# 블로그, 뉴스, 책 한꺼번에 가져오기

```python
search = input('책을 검색해주세요')
```

```python
all_data3 =[]
for category in ['blog', 'book','news']:
    start = 1
    while True :
        url = f"https://openapi.naver.com/v1/search/{category}.json"
        payload = {'query': search ,'display': 100, 'start': start, 'sort': "sim" }
        headers = {"X-Naver-Client-Id" : "ky2xDsaZyifHgJ6pdBwU", "X-Naver-Client-Secret": "3nNHDxyhXR"}
        r = requests.get(url, params=payload, headers=headers)
        data = r.json()
        n_data = data['total']
        total_pages = int(data['total']//100) + 1
        print(r.url)
        print(r.status_code)
        print("전체 검색 테이터수:", n_data, "검색 페이지수:",total_pages)
        resurt_unpack = json_unpack(data)
        all_data3.append(resurt_unpack)
        if start == 1000:
            break
        if start < 901:
            start += 100
        elif start == 901:
            start += 99
final_result3 = pd.concat(all_data3)
final_result3 = final_result3.reset_index(drop=True)
final_result3
                   
```
