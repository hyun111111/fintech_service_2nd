```python
import requests
import pandas as pd
import time
import os
```

```python
def json_unpaking(data):
    bank = {}
    for item in data['body']['items']:
        for key, value in item.items():
            bank.setdefault(key, []).append(value)
    bank = pd.DataFrame(bank)
    return bank
```

```python
def json_unpacking2(data):
    product = {}
    for i in data['response']['body']['items']['item']:
        for key, value in i.items():
            product.setdefault(key,[]).append(value)
    product = pd.DataFrame(product)
    return product
```

```python
from bs4 import BeautifulSoup as bs
```

```python
def xml_unpaking(soup):
    product = {}
    for i in soup.select("item"):
        for item in i:
            product.setdefault(item.name, []).append(item.text)
    del product[None]
    product = pd.DataFrame(product)
    return product    
```

# key_list를 만들고 select로 직접 추출하기

```python
df_list = []
page = 1
while True :
    url = "https://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    key = "bmvKYhJ5bZui/zXZQuUKx+C2Ma+hmR/x/pPGfHBoB17yLKB342mU66pZd0WyvB0P28Cxj0jFB/tiXgVZlmZJHg=="
    payload = dict(serviceKey=key,numOfRows=1000, pageNo=page ,resultType="xml")
    r = requests.get(url,params=payload)
    print(r.url, r.status_code)
    data = r.text
    soup = bs(data,"xml")
    total = int(soup.select_one("totalCount").text)
    num= int(soup.select_one("numOfRows").text)
    key_list = {}
    for item in soup.select("*"):
        key_list.setdefault(item.name,"")
    key_list = list(key_list.keys())
    result = {}
    for key in key_list[10:]:
        for item in soup.select(key):
            result.setdefault(item.name,[]).append(item.text)
    df_list.append(pd.DataFrame(result))
    if page*num < total:
        page += 1
    else : 
        break
# 인덱스 해서 뽑는게 더 빠르다. 저 * 가 오래걸리는 주범
```

```python
result_df = pd.concat(df_list)
result_df = result_df.reset_index(drop=True)
result_df
```
