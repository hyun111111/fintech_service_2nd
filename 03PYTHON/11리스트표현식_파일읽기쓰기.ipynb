# 리스트 표현식
* 리스트 안에 for 반복문과 if else 조건문을 한번에 적어서 리스트를 만드는 방법
* 여러 줄의 코드를 1줄로 줄일 수 있다.
* 속도가 느린 파이썬의 단점을 보완하기 위함
* for문과 if else 문을 사용해서 처리할때보다 속도가 빠르다.
* [실행할 명령 for 변수 in list, tuple]

```python
for i in range(1,11):
    if i%2 == 0 :
        print(f"{i}는 짝수입니다")
    else:
        print(f"{i}는 홀수입니다")
```

```python
for i in range(1,11):
        print(f"{i}는 짝수입니다") if i%2 == 0 else print(f"{i}는 홀수입니다")
```

## 리스트 표현식과 일반 반복문의 속도 차이

```python
import time
```

```python
start = time.time()
a = []
for i in range(1,200000000):
    a.append(i+5)
print(a[:10])

print("코드실행시간:", time.time()- start)
```

```python

start2 = time.time()
b = [i+5 for i in range(1,200000000)]
print(b[:10])
print("코드실행시간:", time.time()-start2)
```

```python
import numpy as np
```

```python
start3 = time.time()

c = np.arange(1,200000000)
c = c+5
print(c[:10])
print("코드실행시간:", time.time()-start3)
```

## with open

```python
with open("text2.txt", "w") as file3:
    file3.write("with open을 사용하면 파일을 자동으로 닫아주어 편리하다")
```

```python
with open ("text2.txt", "r") as file4:
    data = file4.read()
print(data)
```
