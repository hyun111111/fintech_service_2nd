```python
a_reviews = {}
for page in data:
    for item in page['data']['body']["reviewContent"]["reviews"]["hermes"]["groups"]:
        for dicts in item['items']:
            for key, value in dicts.items():
                if key == 'reviewDate':
                    a_reviews.setdefault(key,[]).append(value)
                elif key == 'itineraryId' :
                    a_reviews.setdefault(key,[]).append(value)
                elif key == 'tripType':
                    a_reviews.setdefault(key,[]).append(value)
                elif key == 'rating' :
                    a_reviews.setdefault(key,[]).append(value)
                elif key == 'description' :
                    a_reviews.setdefault(key,[]).append(value)

print(len(a_reviews['reviewDate']))
```

```python
import pandas as pd

# 모든 리스트의 길이를 최장 길이로 맞춤
max_length = max(len(v) for v in a_reviews.values())

# 리스트 길이가 짧은 경우 None을 채워줌
for key in a_reviews:
    a_reviews[key] += [None] * (max_length - len(a_reviews[key]))

a_reviews_dp = pd.DataFrame(a_reviews)
```
