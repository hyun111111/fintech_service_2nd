```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib 
import seaborn as sns
```

```python
df = pd.read_csv("./data/salary2.csv")
df.head()
```

```python
df2 = df.copy()
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df['class'] = df['class'].apply(lambda x:1 if x==' >50K' else 0)
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop('class', axis=1)
y =  df['class']
```

```python
X
```

```python
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4, stratify=y, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dtc = DecisionTreeClassifier(random_state =10)
dtc.fit(X_train, y_train)
pred = dtc.predict(X_test)
print(classification_report(y_test, pred))
```

# 클래스 불균형을 처리하는 방법

* 연봉 데이터의 경우 >50k가 1, <=50k사 3의 비율을 가지고 있음,
* train_test_splittl stratify=옵션을 반드시 포함
* 머신러닝 알고리즘에 있는 소수 클래스에 가중치를 주는 옵션 on

# k-fold, 하이퍼하라미터 튜닝
* k-fold 교차검증: 훈련에 사용할 데이터를 여러개로 나눠서 편향을 줄이고 균형잡히게 하는 방법
* 하리퍼파라미터 튜닝 : 머신러닝 알고리즘에 있는 여러 파라미터를 조정해서 최고이 성능을 내는 모델을 만드는 것, mas_depth, n_estimators, learning _rate
* grid.search, random.search
* gridsaerch: 지정해놓은 모든 경우의 수의 하이퍼 파라미터를 탐색
* randomsearch: 지정해놓은 범위 내에서 랜덤하게 탐색

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state =10)
```

```python
params = dict(criterion=["gini", "entropy", "log_loss"], 
              max_depth=[1,5,10,15], 
              random_state =[7,10,777],
              class_weight=[None,"balanced"]
             )
```

```python
grid_cv = GridSearchCV(model, param_grid=params,cv=3, n_jobs=-1)
grid_cv.fit(X_train,y_train)
pred = grid_cv.predict(X_test)
print("best_params: ", grid_cv.best_params_)
print("best_score_: ", grid_cv.best_score_)
```

```python
model2 = DecisionTreeClassifier(class_weight= None, criterion='gini', max_depth= 10, random_state=777)
model2.fit(X_train,y_train)
pred = model2.predict(X_test)
print(classification_report(y_test, pred))
```

```python
params = dict(criterion=["gini", "entropy", "log_loss"], 
              max_depth=[8,9,10,11,12], 
              random_state =[777],
              class_weight=[None,"balanced"]
             )
```

```python
grid_cv = GridSearchCV(model, param_grid=params,cv=3, n_jobs=-1)
grid_cv.fit(X_train,y_train)
pred = grid_cv.predict(X_test)
print("best_params: ", grid_cv.best_params_)
print("best_score_: ", grid_cv.best_score_)
```

gridSearch에서 score기준 변경하기
* scoring =[accuracy,precision, recall, f1, roc_auc]

```python
grid_cv = GridSearchCV(model, param_grid=params,cv=3, scoring='recall', n_jobs=-1)
grid_cv.fit(X_train,y_train)
pred = grid_cv.predict(X_test)
print("best_params: ", grid_cv.best_params_)
print("best_score_: ", grid_cv.best_score_)
```

```python
model3 = DecisionTreeClassifier(class_weight= 'balanced', criterion='gini', max_depth= 10, random_state=777)
model3.fit(X_train,y_train)
pred = model3.predict(X_test)
print(classification_report(y_test, pred))
```

# randomSearch
* 지정한 파라미터 값에서 random한 값을 더해서 최적 값을 찾아냄

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
dtc = DecisionTreeClassifier(random_state =10)
```

```python
params = dict(criterion=["gini", "entropy", "log_loss"], 
              max_depth=[1,5,10,15], 
              random_state =[7,10,777],
              class_weight=[None,"balanced"]
             )
```

```python
rand_cv = RandomizedSearchCV(dtc , param_distributions=params, scoring='f1', cv=3, n_jobs=-1, random_state=10)
rand_cv.fit(X_train,y_train)
pred = rand_cv.predict(X_test)
print("best_params: ", rand_cv.best_params_)
print("best_score_: ", rand_cv.best_score_)
```

```python
model3 = DecisionTreeClassifier(class_weight= 'balanced', criterion='entropy', max_depth= 15, random_state=10)
model3.fit(X_train,y_train)
pred = model3.predict(X_test)
print(classification_report(y_test, pred))
```

XGBoost를 랜덤서치로 하이퍼파라미터 튜닝하기

```python
xgb = XGBClassifier(n_jobs=-1,  random_state=10)
```

```python
xgb_params = dict(n_estimators=[100,500,1000],
                  max_depth=[1,5,10,15,20],
                  leaning_rate=[0.1,0.5,1],
                  booster=['gbtree', 'gblinear','dart'],
                  scale_pos_weight = [0,3]
                 )
                  
```

```python
rand_cv = RandomizedSearchCV(xgb, param_distributions=xgb_params, scoring='roc_auc', cv=3,  random_state=10)
rand_cv.fit(X_train,y_train)
pred = rand_cv.predict(X_test)
print("best_params: ", rand_cv.best_params_)
print("best_score_: ", rand_cv.best_score_)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
xgb2 = XGBClassifier(n_jobs=-1,  random_state=10, 
                     scale_pos_weight= 3, n_estimators= 1000, max_depth= 1, leaning_rate= 0.5, booster= 'gbtree')
xgb2.fit(X_train,y_train)
pred = xgb2.predict(X_test)
print(classification_report(y_test, pred))
print(roc_auc_score(y_test, pred))
```
