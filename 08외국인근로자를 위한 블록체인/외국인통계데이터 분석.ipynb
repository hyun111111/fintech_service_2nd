```python

# 1) CSV 파일 불러오기 (인코딩은 파일에 맞게 조정)
file_path = './data/한국국적을_가지지_않은_자외국인_근로자_남__20250624113941.csv'
# 만약 utf-8-sig 로 읽어오지 못하면, encoding='cp949' 로 시도하세요.
df = pd.read_csv(file_path)
print(df)
# 2) 필요한 컬럼만 선택 (‘국적’과 ‘계’)
df_totals = df[['국적', '계']]
print(df)
# 3) 합계·소계·기타 등 요약 행 제거 (원하시는 행 이름을 추가로 제외할 수 있습니다)
exclude = ['합계', '소계', '기타']
df_totals = df_totals[~df_totals['국적'].isin(exclude)]

# 4) ‘계’ 기준 내림차순 정렬
df_sorted = df_totals.sort_values(by='계', ascending=False).reset_index(drop=True)

# 5) 결과 출력
print(df_sorted)
```
